```python
import librosa # for audio analysis
import librosa.display
import soundfile as sf # for reading and writing sound files
import matplotlib.pyplot as plt
import IPython.display as ipd # to display interactive rich media like audios
import numpy as np
```

```python
audio5_1 = "5_yweweler_2.wav"
audio5_2 = "5_yweweler_11.wav"
audio5_3 = "5_yweweler_10.wav"
audio9_1 = "9_yweweler_4 (1).wav"
audio9_2 = "9_yweweler_5.wav"
audio9_3 = "9_yweweler_6.wav"
```

Loading the audio files using librosa.

```python
# spoken 5 audio
y1, sr1 = librosa.load(audio5_1) # y1 = np array of audio signals; sr = sample rate (no. of data points (samples) played back each second -- Hz)
y2, sr2 = librosa.load(audio5_2)
y3, sr3 = librosa.load(audio5_3)
#spoken 9 audio
y4, sr4 = librosa.load(audio9_1)
y5, sr5 = librosa.load(audio9_2)
y6, sr6 = librosa.load(audio9_3)
```

```python
# Play audio files
print("Audio 5_1:")
ipd.Audio(y1, rate=sr1)
```

```python
# Play audio files
print("Audio 9_1:")
ipd.Audio(y4, rate=sr4)
```

# Zero Crossing Rate

```python
zcr1 = librosa.feature.zero_crossing_rate(y1)
print(zcr1)
print('max:',zcr1.max())
print('min:',zcr1.min())
```

```python
zcr2 = librosa.feature.zero_crossing_rate(y2)
print('max:',zcr2.max())
print('min:',zcr2.min())
```

```python
zcr3 = librosa.feature.zero_crossing_rate(y3)
print('max:',zcr3.max())
print('min:',zcr3.min())
```

```python
zcr4 = librosa.feature.zero_crossing_rate(y4)
print('max:',zcr4.max())
print('min:',zcr4.min())
```

```python
zcr5 = librosa.feature.zero_crossing_rate(y5)
print('max:',zcr5.max())
print('min:',zcr5.min())
```

```python
zcr6 = librosa.feature.zero_crossing_rate(y6)
print('max:',zcr6.max())
print('min:',zcr6.min())
```

Observng the above zero-crossing rate values for each audio file, we can observe a pattern that audio files 1,2,3 [spoken 5] have similar ranges of zero-crossing rates. They have relatively higher values, while audio files 4,5,6 [spoken 9] also have  a similar range of zero-crossing rates. But The maximum and minimum values are relatively lower. As this feature somewhat classifies the two audio groups, so this can be considered further for classification model.

# Melspectrogram

```python
mel_signal1 = librosa.feature.melspectrogram(y=y1, sr=sr1, hop_length=2, n_fft=100)
spectrogram = np.abs(mel_signal1)
power_to_db = librosa.power_to_db(spectrogram, ref=np.max)
plt.figure(figsize=(8, 4))
librosa.display.specshow(power_to_db, sr=sr1, x_axis='time', y_axis='mel', cmap='magma', hop_length=2)

plt.colorbar(label='dB')
plt.title('Mel-Spectrogram (dB)', fontdict=dict(size=18))
plt.xlabel('Time', fontdict=dict(size=15))
plt.ylabel('Frequency', fontdict=dict(size=15))
plt.show()
```

```python
mel_signal1.mean()
```
